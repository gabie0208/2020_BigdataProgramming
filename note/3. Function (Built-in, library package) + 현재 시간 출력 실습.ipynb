## 1. 함수란 무엇인가?

- 함수 : 특정의 기능을 수행하는 코드의 집합
1. 기본 함수 / built-in 함수 : 파이썬 실행과 동시에 사용 가능한 함수들
2. 라이브러리 / 패키지 함수 : 해당 라이브러리 포함한 후 사용 가능

## dir 함수

- dif([object]) : dir에 어떠한 object(객체, 함수, 클래스, 모듈, 패키지 등등)가 인자로 넘어오는가에 따라서 다양한 결과가 *list* 타입으로 산출

3. dir(클래스) : 클래스가 가진 속성
4. dir(모듈) : 모듈이 포함하고 있는 요소의 이름
5. dir(패키지) : 패키지가 포함하고 있는 요소의 이름

```python
dir()
```

## sorted 함수

- sorted(반복 가능한 데이터, key=None, reverse = False) -> 정렬된 리스트
- sorted 함수는 반복 가능한 데이터를 입력 받아서 정렬된 리스트로 반환
- key는 정렬한 기준을 넘겨줌. 만약 정렬기준이 없으면 기본 정렬 기준 이용
- reverse는 역순 정렬 시킬 것인지의 여부를 알려주는 인자

```python
numbers = [2, 3, 1, 5, 4]
```

```python
sorted_numbers = sorted(numbers)
sorted_numbers
```

```python
reverse_sorted_numbers = sorted(numbers, reverse = True)
reverse_sorted_numbers
```

```python
numbers
```

sorted 다른 예제

```python
colors = ["red", "green", "blue", "yellow"]
sorted_colors = sorted(colors)
sorted_colors
```

```python
len_sorted_colors = sorted(colors, key=len)
len_sorted_colors
```

```python
reverse_len_sorted_colors = sorted(colors, key=len, reverse=True)
reverse_len_sorted_colors
```

## 3. 라이브러리(패키지) 함수

- 파이썬에서는 함수들을 기능별로 패키지 혹은 모듈에 포함시켜 제공
- 모듈 : 파이썬 코들르 포함하고 있는 파일
- 모듈을 묶어서 패키지 라고 한다
- 파이썬을 설치하면 많은 라이브러리(패키지)와 모듈들이 설치됨
- 각 라이브러리 및 모듈 마다 많은 수의 함수들이 포함
- 기본 라이브러리에서 제공되지 않는 것은 추가의 라이브러리(numpy, scipy, matplotlib etc..)를 설치해서 사용
- 라이브러리나 모듈 내에 포함된 함수를 사용하기 위해서는 import 해야 함

## import 라이브러리명 혹은 모듈명

- 라이브러리나 모듈을 포함시킴
- 대상체로 사용할 경우 -> 라이브러리명 혹은 모듈명.대상체 형태로 사용

```python
import math
math.sin(math.pi)
```

```python
math.log(math.e)
```

## import 라이브러리명 혹은 모듈명 as 축약이름

- 라이브러리나 모듈을 축약된 이름으로 바꾸어서 포함시킴
- 모듈 내의 대상체를 사용할 경우 축약이름.대상체 형태로 사용

```python
import math as ma
ma.sin(ma.pi)
```

```python
ma.log(ma.e)
```

## from 라이브러리명 혹은 모듈명 import 대상체명

- 라이브러리나 모듈로부터 몇몇 대상체만을 포함시킴
- 대상체의 사용은 대상체명을 이용

```python
from math import sin, log, pi, e
sin(pi)
```

```python
log(e)
```

## from 라이브러리명 혹은 모듈명 import *

- 라이브러리나 모듈 내에 포함된 모든 대상체 포함
- 대상체의 사용은 대상체명을 이용

```python
from math import*
sin(pi)
```

```python
log(e)
```
