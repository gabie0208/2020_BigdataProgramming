```python
import numpy as np
import matplotlib.pylab as plt

X = np.linspace(-np.pi,np.pi,256)
S = np.cos(X)

dot = 2*np.pi/3

with plt.xkcd():
    plt.plot(X, S, label = "cos")
    plt.title("XKCD style plot!!!")
    
    plt.yticks([-1.00,-0.75,-0.5,-0.25,0,0.25,0.5,0.75,1])
    plt.scatter([dot],[np.cos(dot)], color = "r")
    
    plt.annotate(r"$0.5 Here!$", xy=(dot,np.cos(dot)), xycoords="data", 
                 xytext=(0.5,0.05),textcoords="axes fraction", fontsize=16,
                 arrowprops=dict(arrowstyle="->", linewidth=1, color="k"))
    plt.show()
```
