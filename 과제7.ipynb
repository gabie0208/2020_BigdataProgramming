```python
class Company:
    name = "ABC Global Company"
    
    @classmethod
    def message(cls):
        print('The Company Name is %s !' %cls.name)
        cls.func_msg()
 
    @staticmethod
    def func_msg():
        print("Welcome to my company !")
  
Company.message()
    
class Person:
    def __init__(self, name, id):
        self._name = name
        self._id = id
        self._age = 0
        
        @property
        def age(self):
            return self._age
        
        @age.setter
        def age(self, value):
            self._age = value
            
CEO = Person('Richard', '0')
Director = Person('Emily', '0')
CEO.age = 65
Director.age = 60
print("CEO 나이: %d" %CEO.age)
print("Director 나이: %d" %Director.age)            
 
class Employee(Person):
    def __init__(self, name, id, org):
        Person.__init__(self, name, id)
        self._org = org
    
    def information(self):
        print("Name: " + self._name + " ID: " + self._id + " Org: " + self._org)
        
class Manager(Employee):
    def __init__(self, name, id, org, salary):
        Employee.__init__(self, name, id, org)
        self._salary = salary
        
    def information(self):
        Employee.information(self)
        print("Salary: " + self._salary + "$")
          
class Staff(Employee):
    def __init__(self, name, id, org, team):
        Employee.__init__(self, name, id, org)
        self._team = team
    
    def information(self):
        Employee.information(self)
        print("Team: " + self._team)
           
class Hourly(Employee):
    def __init__(self, name, id, org, hours):
        Employee.__init__(self, name, id, org)
        self._hours = hours
    
    def information(self):
        Employee.information(self)
        print("Hours: " + self._hours)

manager1 = Manager("Robin", "1", "sales", "50,000")
manager2 = Manager("Gabie", "2", "R&D", "60,000")
manager3 = Manager("Maria", "3", "Management", "45,000")

staff1 = Staff("Joseph", "4", "Management", "Team1")
staff2 = Staff("Lucy", "5", "sales","Team2")
staff3 = Staff("Jenny", "6", "HR","Team3")

hourly1 = Hourly("James", "7", "Entertainment", "50")
hourly2 = Hourly("Lucy", "8", "Entertainment","55")
hourly3 = Hourly("Jenny", "9", "Entertainment","60")
```

```python
import time

def checkTime(func):
    
    def newFunc(*args, **kwargs):
        t = time.strftime('현재 시간 : [%Y-%m-%d %H:%M]', time.localtime(time.time()))
        print(t)
        start = time.time()
        func(*args, **kwargs)
        end = time.time()
        print("\n실행 시간 : ", end - start)
    return newFunc

@checkTime
def aFunc():
    for i in range(1, 101):
        print(end = ' ')

aFunc()
print("\n")

print("--- Manager Name List ---")
manager1.information()
manager2.information()
manager3.information()
print("\n")

print("--- Staff Name List ---")
staff1.information()
staff2.information()
staff3.information()
print("\n")

print("--- Hourly Name List ---")
hourly1.information()
hourly2.information()
hourly3.information()
```
